```python
def conv_province(i) :
    prov = i.split(" ")
    prov = prov[0]
    print(prov)
    if len(prov) == 4 and prov[-1] == "도" :
        return prov[0] + prov[2]
    else :
        return prov[0:2]
    
```

```python
people_path = "D:\\python\\201612_201612_주민등록인구및세대현황_연간.csv"
```

```python
dr = pd.read_csv(people_path, engine="python")
```

```python
dr["그룹명"] = dr.행정구역.agg(conv_province)
```

```python
dr1 = dr.iloc[1:-1][["그룹명", "2016년_총인구수"]]
dr1.columns = ["그룹명", "all"]
```

```python
fdf = dr1.merge(dfb, on="그룹명")
```

```python
fdf["all"] = fdf["all"].agg(lambda x : x.replace(",","").astype(int))
fdf[8] =  fdf[8].astype(int)
```

```python
fdf["result"] = fdf["all"]/fdf[8] * 1000
```
